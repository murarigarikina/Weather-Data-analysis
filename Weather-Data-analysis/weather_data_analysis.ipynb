# Data Visualization:

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

We can observe that wind and preception are weakly co-related.temp_max and wind are negatively correlated means they move in the opposite direction.

## Key Insights
- Maximum and minimum temperatures are strongly correlated
- Wind speed shows weak correlation with precipitation
- Rain is the most frequent weather condition in the dataset
